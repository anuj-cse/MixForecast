```python
def fine_load_results(datasets):
    results = []
    for dataset in datasets:
        print(dataset)
        df = pd.read_csv(f'{dataset}/result.csv')
        df.insert(0, 'dataset', dataset)            
        results.append(df)  
        print(df.shape)
    results_df = pd.concat(results)
    return results_df
```
